```python
import gsd
import gsd.hoomd
import math
t = gsd.hoomd.open('npt.gsd', 'rb');
```

```python
print("number of frames:",len(t))
```

```python
N=t[len(t)-1].particles.N
print("Number of particles:",N)
```

```python
bonds=t[len(t)-1].bonds.N
print("Number of bonds:",bonds)
```

```python

bondsgroup=t[len(t)-1].bonds.group
print(bondsgroup[:40])
```

```python
bondstype=t[len(t)-1].bonds.types
print(bondstype)
```

```python
numberofpolymers=N-bonds
print("number of polymers:",numberofpolymers)
```

```python
monomersineachpolymer=N//numberofpolymers
print("monomers in each polymer:",monomersineachpolymer)
```

```python
position=t[len(t)-1].particles.position
print(len(position))
```

```python
x2, y2, z2, Rg2,index=0,0,0,0,0
xcom, ycom, zcom=0,0,0
arrayRgforeachpolymer=[]


for j in range (numberofpolymers):
     index=j*monomersineachpolymer
     x2, y2, z2, Rg2=0,0,0,0
     xcom, ycom, zcom=0,0,0
        
     for i in range (monomersineachpolymer):
         x2, y2, z2, Rg2=0,0,0,0
         xcom+= float(position[i+index][0])
         ycom+= float(position[i+index][1])
         zcom+= float(position[i+index][2])
         i=i+1
     xcom=xcom/monomersineachpolymer
     ycom=ycom/monomersineachpolymer
     zcom=zcom/monomersineachpolymer

     for k in range (monomersineachpolymer):

        x2+= float((position[k+index][0]-xcom)**2)
        y2+= float((position[k+index][1]-ycom)**2)
        z2+= float((position[k+index][2]-zcom)**2)
        k=k+1
     Rg2+=(x2+y2+z2)

     arrayRgforeachpolymer.append(math.sqrt(Rg2/monomersineachpolymer))
     j=j+1


print(arrayRgforeachpolymer)
```

```python
print(len(arrayRgforeachpolymer))
print(max(arrayRgforeachpolymer))
print(min(arrayRgforeachpolymer))

```
